```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
df.info()
```

```python
x = df["YearsExperience"].values.reshape(-1,1)
y = df["Salary"].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
print(linear_model.coef_)
print(linear_model.intercept_)
a = linear_model.coef_
b = linear_model.intercept_
print("Persamaan linearnya adalah")
print("Y = "+str(a)+" + "+ str(b)+"X")
```

```python
linear_model.score(x_test,y_test)
```

```python
price_predict = linear_model.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, price_predict, c='r')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Plot Experience vs Salary')
```

```python
# menambahkan regulasi
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1 , normalize=True)
ridge.fit(x_train,y_train)
ridge_predict = ridge.predict(x_test)
ridge.score(x_test,y_test)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, ridge_predict, c='y')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Plot Experience vs Salary')
```

```python
#menambah regulasi
from sklearn.linear_model import Lasso
# define model
model = Lasso(alpha=1.0)
model.fit(x_train,y_train)
model.predict(x_test)
model.score(x_test,y_test)
```

```python
l_predict = model.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, l_predict, c='k')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.title('Plot horsepower vs price')
```
