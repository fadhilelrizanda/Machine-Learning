```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing

```

```python
df2 = pd.read_csv("avocado.csv", usecols=["AveragePrice",'Total Volume'])
```

```python
df = df2.head(50)
print(df)
```

```python
x = df['Total Volume'].values.reshape(-1,1)
y = df["AveragePrice"].values.reshape(-1,1)

```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1)
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
print(linear_model.coef_)
print(linear_model.intercept_)
a = linear_model.coef_
b = linear_model.intercept_
print("Persamaan linearnya adalah")
print("Y = "+str(a)+" + "+ str(b)+"X")
```

```python
linear_model.score(x_test,y_test)
```

```python
price_predict = linear_model.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, price_predict, c='r')
plt.xlabel('Volume')
plt.ylabel('Price')
plt.title('Plot Volume vs price')
```
